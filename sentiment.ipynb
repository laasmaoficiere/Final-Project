```python
import pandas as pd
```

```python
data_austin = pd.read_csv("Data/Austin_June_24.csv")
austin_reviews = pd.read_csv("Data/reviews_austin.csv")
data_chicago = pd.read_csv("Data/Chicago_June_24.csv")
chicago_reviews = pd.read_csv("Data/reviews_chicago.csv")
data_columbus = pd.read_csv("Data/Columbus_June_24.csv")
columbus_reviews = pd.read_csv("Data/reviews_columbus.csv")
data_dallas = pd.read_csv("Data/Dallas_June_24.csv")
dallas_reviews = pd.read_csv("Data/reviews_dallas.csv")
data_denver = pd.read_csv("Data/Denver_June_24.csv")
denver_reviews = pd.read_csv("Data/reviews_denver.csv")
data_la = pd.read_csv("Data/LA_June_24.csv")
la_reviews = pd.read_csv("Data/reviews_la.csv")
data_ny = pd.read_csv("Data/NY_June_24.csv")
ny_reviews = pd.read_csv("Data/reviews_ny.csv")
data_sf = pd.read_csv("Data/Sanfrancisco_June_24.csv")
data_sandiego = pd.read_csv("Data/SanDiego_June_24.csv")
sandiego_reviews = pd.read_csv("Data/reviews_sandiego.csv")
sf_reviews = pd.read_csv("Data/reviews_sanfrancisco.csv")
data_seattle = pd.read_csv("Data/Seattle_June_24.csv")
seattle_reviews = pd.read_csv("Data/reviews_seattle.csv")
```

```python
#rename id column to listing_id
data_austin.rename(columns={'id':'listing_id'}, inplace=True)
data_chicago.rename(columns={'id':'listing_id'}, inplace=True)
data_columbus.rename(columns={'id':'listing_id'}, inplace=True)
data_dallas.rename(columns={'id':'listing_id'}, inplace=True)
data_denver.rename(columns={'id':'listing_id'}, inplace=True)
data_la.rename(columns={'id':'listing_id'}, inplace=True)
data_ny.rename(columns={'id':'listing_id'}, inplace=True)
data_sf.rename(columns={'id':'listing_id'}, inplace=True)
data_sandiego.rename(columns={'id':'listing_id'}, inplace=True)
data_seattle.rename(columns={'id':'listing_id'}, inplace=True)
```

```python
austin = pd.merge(data_austin, austin_reviews, on='listing_id')
chicago = pd.merge(data_chicago, chicago_reviews, on='listing_id')
columbus = pd.merge(data_columbus, columbus_reviews, on='listing_id')
dallas = pd.merge(data_dallas, dallas_reviews, on='listing_id')
denver = pd.merge(data_denver, denver_reviews, on='listing_id')
la = pd.merge(data_la, la_reviews, on='listing_id')
ny = pd.merge(data_ny, ny_reviews, on='listing_id')
sf = pd.merge(data_sf, sf_reviews, on='listing_id')
sandiego = pd.merge(data_sandiego, sandiego_reviews, on='listing_id')
seattle = pd.merge(data_seattle, seattle_reviews, on='listing_id')
```

```python
#concatenate all the dataframes
data = pd.concat([austin, chicago, columbus, dallas, denver, la, ny, sf, sandiego, seattle])
```

```python
data
```

```python
#take 10% of the data
data = data.sample(frac=0.1)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#drop rows in comments column that are null
data = data.dropna(subset=['comments'])
```

```python
from textblob import TextBlob

# Perform sentiment analysis
data['sentiment'] = data['comments'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
data['sentiment'].to_csv('Data/sentiment.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Histogram of sentiment scores
plt.figure(figsize=(10, 6))
plt.hist(data['sentiment'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

- Overall Positive Sentiment: The majority of the reviews are skewed towards the positive side of the sentiment scale (with most scores between 0.25 and 0.75), suggesting that most Airbnb listings in your dataset receive favorable reviews.
- Few Negative Reviews: There are relatively few negative reviews (sentiment scores below 0), which is a good indicator for the overall quality of the listings.
- Some Very Positive Reviews: The peak near 1 suggests that some reviews are extremely positive, possibly reflecting very satisfied guests.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='sentiment', y='price', data=data, alpha=0.5)
sns.regplot(x='sentiment', y='price', data=data, scatter=False, color='red')
plt.title('Scatter Plot of Price vs. Sentiment with Regression Line')
plt.xlabel('Sentiment Score')
plt.ylabel('Price')
plt.show()
```

This scatter plot effectively shows that there is no meaningful linear relationship between the sentiment of reviews and the price of Airbnb listings. The sentiment score doesn't appear to influence the price in any significant way, as evidenced by the flat regression line and the distribution of data points.
