```python
import pandas as pd
```

```python
df_no_outliers = pd.read_csv('Data/Cleaned_Airbnb_Data_No_Outliers.csv')
```

```python
df_no_outliers
```

```python
df_no_outliers.keys()
```

```python
#drop columns that are not needed
df_no_outliers = df_no_outliers.drop(['host_id', 'latitude', 'longitude','price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365', 'number_of_reviews_ltm' ], axis=1)
```

```python
#drop room_type column
df_no_outliers = df_no_outliers.drop(['room_type'], axis=1)
```

```python
df_no_outliers
```

```python
#check uique states
df_no_outliers['state'].unique()
```

```python
# Drop duplicate neighborhoods within the same city
df_unique_neighborhoods = df_no_outliers.drop_duplicates(subset=['city', 'neighbourhood'])

# Display the resulting DataFrame
print(df_unique_neighborhoods)
```

```python
#change order of columns - state, city, neighbourhood
df_unique_neighborhoods = df_unique_neighborhoods[['state', 'city', 'neighbourhood']]
```

```python
# Save the resulting DataFrame to a new CSV file
df_unique_neighborhoods.to_csv('Data/Unique_Neighborhoods.csv', index=False)
```

```python
# Check for duplicate neighborhoods within each city
duplicates = df_unique_neighborhoods[df_unique_neighborhoods.duplicated(subset=['city', 'neighbourhood'], keep=False)]

# Display the duplicates
print(duplicates)
```
