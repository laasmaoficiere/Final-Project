# Importing and Merging Data

```python
import pandas as pd
```

#### Austin

```python
austin_1 = pd.read_csv('Data/Austin_September_23.csv')
austin_2 = pd.read_csv('Data/Austin_December_23.csv')
austin_3 = pd.read_csv('Data/Austin_March_24.csv')
austin_4 = pd.read_csv('Data/Austin_June_24.csv')
```

```python
# List of DataFrames
dataframes = [austin_1, austin_2, austin_3, austin_4]

# Corresponding year, month, city, and state for each DataFrame
dates = [(2023, 'September'), (2023, 'December'), (2024, 'March'), (2024, 'June')]
city_name = "Austin"  # Set the city name
state_name = "Texas"  # Set the state name

# Loop through each DataFrame and corresponding date
for df, (year, month) in zip(dataframes, dates):
    df['year'] = year
    df['month'] = month
    df['city'] = city_name  # Add the city name column
    df['state'] = state_name  # Add the state name column

# Concatenate the DataFrames
austin_combined = pd.concat(dataframes).reset_index(drop=True)

# Save the combined DataFrame to a CSV file
austin_combined.to_csv('Data/Austin_Combined.csv', index=False)
```
